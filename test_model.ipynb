```python
import ultralytics
from ultralytics import YOLO
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#testing a video
cap = cv2.VideoCapture(r"C:\Users\kbtha\Desktop\Mask Detection\Test_video3.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
writer = cv2.VideoWriter(r"C:\Users\kbtha\Desktop\Mask Detection\output _video3.mp4",cv2.VideoWriter_fourcc(*'DIVX'),25,(width,height))

# Load a model
model = YOLO(r"C:\Users\kbtha\Desktop\Mask Detection\best.pt")

threshold = 0.38

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1, cv2.LINE_AA)

    writer.write(frame)
    cv2.imshow("Detection", frame)
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

cap.release()
writer.release()
cv2.destroyAllWindows()
```
